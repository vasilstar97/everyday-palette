```python
from everyday_palette.processing import Processor
from everyday_palette.generating import Generator

processor = Processor.load('scaler.joblib')
generator = Generator.load('gmm.joblib')
```

```python
def generate_palette():
    while True:
        samples = generator.sample(1)
        palette = processor.inverse_transform(samples)[0]
        if palette.is_above_threshold():
            return palette
```

```python
palette = generate_palette()
palette.to_img(1024)
```
